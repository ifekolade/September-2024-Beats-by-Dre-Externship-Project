## 6. Sentiment Analysis

### 6.1 Methodology


For the sentiment analysis, textblob was used to assign sentiment scores to each of the reviews. Additionally, it used these scores to classify the reviews into negative and positive reviews. There was also the option of neutral review.

This is an example using sample reviews


```python
# Positive, Negative and Neutral Sentiment
import pandas as pd
from textblob import TextBlob

reviews = [
    "Great product! I love it!",
    "It's okay, but not what I expected.",
    "Very disappointed with this product.",
    "Excellent quality and value for money.",
    "I would not recommend this product.",
    "It's a good product, but a bit overpriced.",
    "I'm very satisfied with this purchase.",
    "It's a great product for the price.",
    "I'm not sure if I would buy it again.",
    "I'm very happy with this product",
]

sample_df = pd.DataFrame({'reviews': reviews})

def classify_sentiment(text):
    blob = TextBlob(text)
    sentiment_polarity = blob.sentiment.polarity

    if sentiment_polarity > 0:
        return 'Positive'
    elif sentiment_polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

sample_df['sentiment'] = sample_df['reviews'].apply(classify_sentiment)

print(sample_df)
```

FURTHER ANALYSIS OF REVIEW SENTIMENT OVER TIME

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load the dataset
df_beats = pd.read_csv('reviews_3.csv')

# Check the column names
print(df_beats.columns)

# Extract the date from the 'date' column
df_beats['timestamp'] = df_beats['timestamp'].str.split('United States ').str[1]

# Convert the 'date' column to datetime objects
df_beats['timestamp'] = pd.to_datetime(df_beats['timestamp'])

# Function to calculate sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Calculate sentiment polarity for each review
df_beats['sentiment'] = df_beats['content'].apply(get_sentiment)

# Group the reviews by month and calculate the average sentiment
monthly_sentiment = df_beats.groupby(pd.Grouper(key='timestamp', freq='M'))['sentiment'].mean()

# Create a line plot to visualize the trend of sentiment over time
plt.figure(figsize=(12, 6))
plt.plot(monthly_sentiment.index, monthly_sentiment.values)
plt.xlabel('Date')
plt.ylabel('Average Sentiment')
plt.title('Sentiment Trend for Beats Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
